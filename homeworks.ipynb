```python
# from pylab import *
# mpl.rcParams['font.sans-serif'] = ['SimHei']
# plt.rcParams['axes.unicode_minus']=False

theta = np.linspace(-np.pi,np.pi,256)
x = 6*np.sin(theta) +1
y = 5*np.cos(theta)
ax = plt.gca()
ax.set_aspect(1)
ax.spines['right'].set_color('none')  # None ok
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',1))
ax.spines['bottom'].set_position('zero')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False   # 
plt.title('')

plt.yticks([])
plt.xticks([-5,0,5])
plt.fill_between(x,0,y,(x<1) & (y>0))
plt.plot(x,y)
```

```python
def draw_cricle(r=1,x=0,y=0):
    theta = np.linspace(-np.pi,np.pi,256)
    x = r*np.sin(theta)+x
    y = r*np.cos(theta)+y
    ax = plt.gca()
    ax.set_aspect(1)
    plt.plot(x,y)
    return plt,x,y
draw_cricle(3)

```

```python
def set_someting(plt,x=0):
    ax=plt.gca()
    ax.spines['right'].set_color('none')  # None ok
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_position(('data',x))
    ax.spines['bottom'].set_position('zero')
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus']=False   # 
    plt.title('')
```

```python
plt1 = draw_cricle()
set_someting(plt1)
```

```python
def my_fill(plt,x,y1,y2,condition):
    plt.fill_between(x,y1,y2,condition)
    
plt1,x1,y1 = draw_cricle(r=10)
    
my_fill(plt1,x1,0,y1,y1>0)
```
