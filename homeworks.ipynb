# 三题

第三题: 写个方法draw_circle,画个圆,
要求:
1.圆以原点为中心,可以改变圆的大小,
2. 图例显示:圆
3. 上下左右四个轴,去掉上右两个轴
4. 将 左下两个轴移到中间
5. 将y轴的刻度去掉, 将x轴的刻度设置为[-5,0,5]


```python
# from pylab import *
# mpl.rcParams['font.sans-serif'] = ['SimHei']
# plt.rcParams['axes.unicode_minus']=False

theta = np.linspace(-np.pi,np.pi,256)
x = 6*np.sin(theta) +1
y = 5*np.cos(theta)
ax = plt.gca()
ax.set_aspect(1)
ax.spines['right'].set_color('none')  # None 也ok
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',1))
ax.spines['bottom'].set_position('zero')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False   # 显示负号
plt.title('中文')

plt.yticks([])
plt.xticks([-5,0,5])
plt.fill_between(x,0,y,(x<1) & (y>0))
plt.plot(x,y)
```

```python
def draw_cricle(r=1,x=0,y=0):
    theta = np.linspace(-np.pi,np.pi,256)
    x = r*np.sin(theta)+x
    y = r*np.cos(theta)+y
    ax = plt.gca()
    ax.set_aspect(1)
    plt.plot(x,y)
    return plt,x,y
draw_cricle(3)

```

```python
def set_someting(plt,x=0):
    ax=plt.gca()
    ax.spines['right'].set_color('none')  # None 也ok
    ax.spines['top'].set_color('none')
    ax.spines['left'].set_position(('data',x))
    ax.spines['bottom'].set_position('zero')
    plt.rcParams['font.sans-serif'] = ['SimHei']
    plt.rcParams['axes.unicode_minus']=False   # 显示负号
    plt.title('中文')
```

```python
plt1 = draw_cricle()
set_someting(plt1)
```

```python
def my_fill(plt,x,y1,y2,condition):
    plt.fill_between(x,y1,y2,condition)
    
plt1,x1,y1 = draw_cricle(r=10)
    
my_fill(plt1,x1,0,y1,y1>0)
```
