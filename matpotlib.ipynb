```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 绘制line

```python
x =np.linspace(-np.pi,np.pi,256,endpoint=True)
```

```python
c,s = np.cos(x),np.sin(x)
```

```python
plt.figure(1)
```

```python
plt.plot(x,c)
```

```python
plt.plot(x,s)
```

```python
plt.title('sin & cos')
```

```python
plt.show()
```
